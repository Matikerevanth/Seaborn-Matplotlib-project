# Handeling the miss values .

```python
# one method for handeling missing values( using inplace ) 
data.fillna(0, inplace=True) # for numeric values 
data.fillna('N/A', inplace=True) # for text values 
```

```python
data.isnull().sum()
```

```python
# method 2: using select_dtypes()-> it is perfect for all the datasets 
# fill numerical columns with 0
numerical_data= data.select_dtypes(include='number').columns
data[numerical_data] = data[numerical_data].fillna(0)
```

```python
# fill text colunms with 'unknow'
categorical_data= data.select_dtypes(include='object').columns
data[categorical_data] = data[categorical_data].fillna('unknow')
```

# 1. Who are the Top Winning Teams 

# Seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# count how many matches each team has won
top_teams = data['winner'].value_counts().head(5).reset_index()
top_teams.columns = ['Team', 'Wins']
# create barplot
plt.figure(figsize= (8,5))
sns.barplot( x= 'Team', y= 'Wins', data=top_teams, palette='viridis')
# add title 
plt.title('Top 5 Winning Teams')
plt.xlabel('Number of wins')
plt.ylabel('Teams')
plt.show()
```

# Matplotlib:


```python
top_teams = data['winner'].value_counts().head(5)
plt.figure(figsize=(8,5))
plt.barh(top_teams.index, top_teams.values, color='orange')
plt.title('Top 5 Winning Teams')
plt.xlabel('Number of Wins')
plt.ylabel('Wins')
plt.gca().invert_yaxis() # Highest team on top
plt.legend(labels = 'lower')
plt.show()
```

# 2. Winning the Toss will help the Winning the Match

# Seaborn:

```python
data['toss_match'] = data['toss_won'] == data['winner']
plt.figure(figsize=(6,4))
sns.countplot(x='toss_match', data=data, palette='Set2')
plt.xticks([0,1], ['no', 'yes'])
plt.title('Toss Winner will also Won the Match')
plt.xlabel('Toss= Match Winner')
plt.ylabel('Match count')
plt.show()
```

```python
 data['home_key_batsman']=data['home_key_bowler'] == data['winner']
plt.figure(figsize=(6,4))
sns.countplot(x='toss_match',data=data, palette='Set2')
plt.xticks([0,1], ['no', 'yes'])
plt.title('Toss Winner will also Won the Match')
plt.xlabel('Key batsman = Match Winner')
plt.ylabel('Match count')
plt.show()
```

# Matplotlib:

```python
toss_match = data['toss_match'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(toss_match, labels=['NO', 'YES'], colors=['red', 'green'])
plt.title('Toss winner will also win match')
plt.show()
```

# 4. Number of Matches per Season

# Seaborn:

```python
season_count = data['season'].value_counts().sort_index().reset_index()
season_count.columns = ['season', 'Matches']
plt.figure(figsize=(10,5))
sns.lineplot(x='season', y='Matches', data=season_count, marker='*', color='blue')
plt.title('Number of Match per Season')
plt.show()
```

# Matplotlib:

```python
season_count = data['season'].value_counts().sort_index()
plt.figure(figsize=(10,5))
plt.plot(season_count.index,season_count.values, marker='o', ls = '-', color = 'green')
plt.title('Number of Matches per season')
plt.xlabel('season')
plt.ylabel('Matches')
plt.show()
```
